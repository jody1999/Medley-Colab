```python
import os
import pypianoroll 
import pretty_midi
```

```python
def process_bar_folder(midi_path, pianoroll_path):
    bars = [f for f in os.listdir(midi_path)]
    for bar in bars[1:]:
        # beat_resolutiono: The number of time steps used to represent a beat.
        try: 
            pianoroll_bar = pypianoroll.parse(os.path.join(midi_path, bar), beat_resolution=24, name=bar)
        except (ValueError,IndexError, FileNotFoundError):
            return
        pypianoroll.save(os.path.join(pianoroll_path,bar) , pianoroll_bar, compressed=True)        
```

```python
def process_singer_folder(singer_path):    

    bar_folders = [f for f in os.listdir(singer_path) if os.path.isfile(os.path.join(singer_path, f)) == False]
    print("Total %3d bar folders in the singer folder" %(len(bar_folders)))
    for bar_folder in bar_folders:        
        pianoroll_folder = r'D:\Music\dataset\piano-roll-bar-cleaned'        
        new_path = os.path.join(pianoroll_folder,bar_folder)
        try:
            os.mkdir(new_path)
        except (OSError):
            print('already exists')
        process_bar_folder(os.path.join(singer_path, bar_folder), new_path)
    print('Folder ' + singer_path +  ' done')
```

```python
subdirs = [x[0] for x in os.walk('D:\Music\dataset\clean_midi_LAKH')][4500:]
for singer_path in subdirs:
    process_singer_folder(singer_path)
```
